```python
caffe_root = '/home/alvin/caffe/caffe-augment/examples'
import os
os.chdir(caffe_root)
import sys
sys.path.insert(0, '/home/alvin/caffe/caffe-augment/python')
import caffe
import cv2
import numpy as np
```

```python
model_proto = '/home/alvin/caffe/caffe-augment/examples/project/classify/mobilenet_v1/deploy.prototxt'
model_weight = '/home/alvin/caffe/caffe-augment/examples/project/classify/mobilenet_v1/models/mobilenet_finetune_iter_3000.caffemodel'
imgtxt = '/home/alvin/caffe/caffe-augment/examples/project/classify/dataset/all_shuffle_val.txt'
testsize = 96
data_set = '/home/alvin/caffe/caffe-augment/examples/project/classify/dataset/'
```

```python
def start_test(model_proto, model_weight, imgtxt, testsize, enable_crop):
    caffe.set_device(0)
    net = caffe.Net(model_proto, model_weight, caffe.TEST)
    imgs = open(imgtxt, 'r').readlines()
    count = 0
    acc = 0
    
    for imgname in imgs:
        imgname, label = imgname.strip().split(' ')
        imgtype = imgname.split('.')[-1]
        if imgtype != 'jpeg' and imgtype != 'jpg' and imgtype != 'JPG' and imgtype != 'png' and imgtype != 'tif' and imgtype != 'bmp':
            print(imgtype, 'error')
            continue
        imgpath = data_set + imgname
        img = cv2.imread(imgpath)
    
        if img is None:
            print("img is empty: ", imgpath)
            continue
        imgheight, imgwidth,channel = img.shape
        
        if enable_crop == 1:
            print("use crop")
            cropx = (imgwidth - testsize)/2
            cropy = (imgheight - testsize)/2
            img = img[cropy:cropy+testsize,cropx:cropx+testsize,0:channel]
        else:
            img = cv2.resize(img, (testsize,testsize),interpolation=cv2.INTER_NEAREST)
            
        transformer = caffe.io.Transformer({'data':net.blobs['data'].data.shape})
        transformer.set_mean('data',np.array([140.008,116.669,122.675]))
        transformer.set_transpose('data',(2,0,1))
        
        out = net.forward_all(data=np.asarray([transformer.preprocess('data',img)]))
        
        result = out['classifier'][0]
        print('result=', result)
        predict = np.argmax(result)
        if str(label) == str(predict):
            acc = acc + 1
        count = count + 1
    print('acc=',float(acc)/float(count))
            
```

```python
start_test(model_proto, model_weight, imgtxt, testsize, 1)
```
